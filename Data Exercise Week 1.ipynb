```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from excel file
# equivalent R code:
# bp_data<-read_excel("bp_2023.xlsx",sheet = "Primary Energy - Cons by fuel",range="A3:O94")

try: 
  bp_data = pd.read_excel(
    "./bp_2023.xlsx", 
    sheet_name="Primary Energy - Cons by fuel", 
    usecols="A:O", 
    skiprows=[*range(2), *range(94, 100)] 
  )
except FileNotFoundError:
  print('File not found!!!')

# print the dataframe
bp_data
```

```python
# clean data to get desired columns
desired_columns = bp_data.columns[[0, *range(8,15)]]
bp_data = bp_data[desired_columns]

# equivalent R code
# bp_data<-bp_data %>% select(1,seq(9,15))

# print out new dataframe
bp_data
```

```python
# change column names
bp_data.columns = ["Country","Oil","Natural Gas","Coal","Nuclear","Hydro","Renewables","Total"]

# equivalent R code
# names(bp_data)<-c("country","Oil","Natural Gas","Coal","Nuclear","Hydro","Renewables","Total")

# print out new dataframe
bp_data
```

```python
# data filtering for NaN values
# remove observations that have an NA for country
bp_data = bp_data.dropna(subset=['Country'])

# equivalent R code
# bp_data<-bp_data%>%filter(!is.na(country))

bp_data
```

```python
# only grab 'totals' data
bp_data = bp_data[bp_data['Country'].str.contains('Total')]

# equivalent R code
# bp_data<-bp_data%>%filter(grepl("Total",country))

bp_data
```

```python
# replace text to strip the word “Total” out of all my regions
bp_data['Country'] = bp_data['Country'].str.replace('Total', '').str.strip()

# equivalent R code
# bp_data<-bp_data%>%mutate(country=gsub("Total ","",country))

# now this is the data that we want to graph
bp_data
```

```python
# pivot the data to a longer style, by country
# Use melt to pivot
bp_data = bp_data.melt(
  id_vars=['Country'], 
  value_vars=['Oil', 'Natural Gas', 'Coal', 'Nuclear', 'Hydro', 'Renewables', 'Total'], 
  var_name='source', 
  value_name='tpes'
)

# equivalent R code

bp_data
```

```python
# take out the global observations
bp_data = bp_data[~bp_data['Country'].str.contains('World')]

# take out the 'total' column
bp_data = bp_data[~bp_data['source'].str.contains('Total')]


# create graph
bp_data = bp_data.pivot(index='Country', columns='source', values='tpes')
ax = bp_data.plot(kind='bar', stacked=True, figsize=(10, 7))
ax.set_ylabel('TPES (Exajoules)')
ax.set_xlabel('country')
ax.set_title('Primary Energy Consumption by Source')
plt.style.use('ggplot')
plt.xticks(rotation=45)
plt.show()
```
